```python
import requests
import json
import pandas as pd
import time

ls_num=[]
session_num=[]
api_data=[]
url=[]

base_url='https://sansad.in/api_ls/member'

lok_sabha_num = 15

while lok_sabha_num <= 18 :
    session =1
    while session <=20 :
        session_url=f"https://sansad.in/api_ls/member/getMemberAttendanceMonthWise?loksabha={lok_sabha_num}&session={session}&locale=en"

        try:
            session_response= requests.get(session_url)
            session_response.raise_for_status()
            sessions= session_response.json()

            ls_num.append(lok_sabha_num)
            session_num.append(session)
            api_data.append(sessions)
            url.append(session_url)
            session += 1
        except requests.exceptions.RequestException as e :
            print({e})


    lok_sabha_num += 1
    time.sleep(1)

    ls_data= pd.DataFrame(
        {
            'lok-sabha':ls_num,
            'session':session_num,
            'api-data':api_data,
            'api-url':url
        }
    )
```

```python
exploded_df=ls_data.explode('api-data').reset_index().drop(columns=['index'])
exploded_df = exploded_df[exploded_df['api-data'].notna()]
exploded_df['dateOfAttendance']=exploded_df['api-data'].apply(lambda x: x['dateOfAttendance'])
exploded_df['dateOfAttendance']=pd.to_datetime(exploded_df['dateOfAttendance'])
exploded_df['dateOfAttendance']=exploded_df['dateOfAttendance'].dt.strftime('%Y/%m/%d')
```

```python
exploded_df.to_csv("exploded.csv")
```

```python
member_df=exploded_df.copy()
```

```python
attendance_result=[]
for index,row in member_df.iterrows() :
  lok_sabha_num=row['lok-sabha']
  session_num=row['session']
  attendance_date=row['dateOfAttendance']
  session_url = f"{base_url}/getMemberAttendanceDateWise?loksabha={lok_sabha_num}&session={session_num}&dateOfAttendance={attendance_date}&locale=en"
  print(f"Lok Sabha: {lok_sabha_num}, Session: {session_num}, Date: {attendance_date}")
  sessions_response = requests.get(session_url)
  sessions_response.raise_for_status()
  sessions = sessions_response.json()
  attendance_result.append(sessions)
  time.sleep(1)

```

```python
member_df.to_csv("member_df.csv",index=False)
```
